# **Finding Seasonality and Trends**

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python

# Perform seasonal decomposition to find trends and seasonality
result = seasonal_decompose(df['price'], model='multiplicative', period=365)
# Plot the decomposed components of the time series
result.plot()
plt.show()
```

# **Plot Observations**

1. The trend component represents the overall pattern in the gold prices, ignoring the noise.
2. The seasonal component shows the repeating patterns or cycles (e.g., yearly fluctuations).
3. The residual component captures the noise or irregularities in the data.

```python
# Perform Augmented Dickey-Fuller test to assess stationarity
adf_result = adfuller(df['price'])
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
for key, value in adf_result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

# **adfuller test result**

The Augmented Dickey-Fuller test is used to check if the time series is stationary. If the p-value is less than 0.05, we reject the null hypothesis and infer that the time series is stationary.

h0=data is not stationary

h1=data is stationary

Here the p_value is greater than 0.05 so we are accepting null hypothesis,i.e data is not stationary.

# **Bringing Dataset in Stationary Form**

```python
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Original Series
fig, (ax1, ax2, ax3) = plt.subplots(3)
ax1.plot(df.price); ax1.set_title('Original Series'); ax1.axes.xaxis.set_visible(False)
# 1st Differencing
ax2.plot(df.price.diff()); ax2.set_title('1st Order Differencing'); ax2.axes.xaxis.set_visible(False)
# 2nd Differencing
ax3.plot(df.price.diff().diff()); ax3.set_title('2nd Order Differencing')
plt.show()
```

### Here we can see how the time series has become stationary. One thing which is noticeable here is in first-order differencing we have fewer noises in the data while after 1st order there is an increase in the noise. So we can select 1st order differencing for our model. We can also verify this using an autocorrelation plot.

```python
df
```

```python
df_diff=df.diff().dropna()
# Perform Augmented Dickey-Fuller test to check if the time series is now stationary
result = adfuller(df_diff['price'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
```

```python


def best_forecasting_model(data, model_params):
    best_aic = np.inf
    best_params = None
    best_model = None

    for params in model_params:
        try:
            # Simple Exponential Smoothing
            if params['model'] == 'simple_exponential':
                model = SimpleExpSmoothing(data).fit(smoothing_level=params['smoothing_level'])
                aic = model.aic

            # Double Exponential Smoothing
            elif params['model'] == 'double_exponential':
                model = ExponentialSmoothing(data, trend=params['trend']).fit()
                aic = model.aic

            # Triple Exponential Smoothing
            elif params['model'] == 'triple_exponential':
                model = ExponentialSmoothing(data, trend=params['trend'], seasonal=params['seasonal'], seasonal_periods=params['seasonal_periods']).fit()
                aic = model.aic
            # Holt-Winters' Exponential Smoothing
            elif params['model'] == 'holt_winters':
                model = ExponentialSmoothing(data, trend=params['trend'], seasonal=params['seasonal'], seasonal_periods=params['seasonal_periods']).fit()
                aic = model.aic

            # ARIMA Models
            elif params['model'] == 'arima':
                model = ARIMA(data, order=params['order']).fit()
                aic = model.aic

            # SARIMA Models
            elif params['model'] == 'sarima':
                model = SARIMAX(data, order=params['order'], seasonal_order=params['seasonal_order']).fit()
                aic = model.aic

            if aic < best_aic:
                best_aic = aic
                best_params = params
                best_model = model
        except: continue

    return best_model,best_params

# Example usage:

```

```python
model_params = [
    {'model': 'simple_exponential', 'smoothing_level': 0.2},
    {'model': 'double_exponential', 'trend': 'add'},
    {'model': 'triple_exponential', 'trend': 'add', 'seasonal': 'add', 'seasonal_periods': 12},
    {'model': 'holt', 'exponential': False, 'smoothing_level': 0.2, 'smoothing_trend': 0.1}, # additive trend
    {'model': 'holt', 'exponential': True, 'smoothing_level': 0.2, 'smoothing_trend': 0.1}, # multiplicative trend
    {'model': 'holt_winters', 'trend': 'add', 'seasonal': 'add', 'seasonal_periods': 12}, # additive trend, additive seasonality
    {'model': 'holt_winters', 'trend': 'add', 'seasonal': 'mul', 'seasonal_periods': 12}, # additive trend, multiplicative seasonality
    {'model': 'holt_winters', 'trend': 'mul', 'seasonal': 'add', 'seasonal_periods': 12}, # multiplicative trend, additive seasonality
    {'model': 'holt_winters', 'trend': 'mul', 'seasonal': 'mul', 'seasonal_periods': 12}, # multiplicative trend, multiplicative seasonality
    {'model': 'arima', 'order': (1,1,0)},
    {'model': 'arima', 'order': (2,1,0)},
    {'model': 'sarima', 'order': (2,1,0), 'seasonal_order': (1,0,0,12)},

    # Simple Exponential Smoothing
    {'model': 'simple_exponential', 'smoothing_level': 0.2},
    {'model': 'simple_exponential', 'smoothing_level': 0.5},
    {'model': 'simple_exponential', 'smoothing_level': 0.8},

    # Holt's Linear Exponential Smoothing
    {'model': 'holt', 'exponential': False, 'smoothing_level': 0.2, 'smoothing_trend': 0.1}, # additive trend
    {'model': 'holt', 'exponential': True, 'smoothing_level': 0.2, 'smoothing_trend': 0.1}, # multiplicative trend
    {'model': 'holt', 'exponential': False, 'smoothing_level': 0.5, 'smoothing_trend': 0.2}, # additive trend
    {'model': 'holt', 'exponential': True, 'smoothing_level': 0.5, 'smoothing_trend': 0.2}, # multiplicative trend

    # Holt-Winters' Exponential Smoothing
    {'model': 'holt_winters', 'trend': 'add', 'seasonal': 'add', 'seasonal_periods': 12}, # additive trend, additive seasonality
    {'model': 'holt_winters', 'trend': 'add', 'seasonal': 'mul', 'seasonal_periods': 12}, # additive trend, multiplicative seasonality
    {'model': 'holt_winters', 'trend': 'mul', 'seasonal': 'add', 'seasonal_periods': 12}, # multiplicative trend, additive seasonality
    {'model': 'holt_winters', 'trend': 'mul', 'seasonal': 'mul', 'seasonal_periods': 12}, # multiplicative trend, multiplicative seasonality

    # ARIMA Models
    {'model': 'arima', 'order': (1,1,0)},
    {'model': 'arima', 'order': (2,1,0)},
    {'model': 'arima', 'order': (1,1,1)},
    {'model': 'arima', 'order': (2,1,1)},

    # SARIMA Models
    {'model': 'sarima', 'order': (2,1,0), 'seasonal_order': (1,0,0,12)},
    {'model': 'sarima', 'order': (1,1,1), 'seasonal_order': (1,0,0,12)},
    {'model': 'sarima', 'order': (2,1,1), 'seasonal_order': (1,0,0,12)},
    {'model': 'sarima', 'order': (1,1,0), 'seasonal_order': (1,1,1,12)},
]


best_model,best_params = best_forecasting_model(df.price, model_params)
```
